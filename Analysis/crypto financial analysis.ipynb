# Investment Multiple and CAGR
two alternative reward metrics that are more intuitive and easier to interpret

```python
multiple=(btc.price[-1]/btc.price[0])  #Investment multiple: Ending value of 1 [Dollar] invested. 
multiple  #multiple=Ending Value/Initial Investment
```

```python
(multiple-1)*100  #Price Increase (in %)
```

```python
btc.price/btc.price[0] #similar/identical concept: Normalized price with base value 1
```

# Drawback of Investment Multiple: Does not take into account investment period. Meaningful only in conjunction with investment period.

# Compound Annual Growth Rate (CAGR): 
The constant annual rate of return that would be required for an investment to grow from its beginning to its ending balance, assuming the profits were invested at the end of each year of investment's lifespan.

```python
start=btc.index[0]
start
```

```python
end=btc.index[-1]
end
```

```python
td=end - start
td
```

```python
td_years=td.days/365.25
td_years
```

```python
cagr=multiple**(1/td_years)-1  #short version
cagr
```

```python
cagr=(btc.price[-1]/btc.price[0])**(1/((btc.index[-1]-btc.index[0]).days/365.25))-1 #long version
cagr
```

```python
(1+cagr)**(td_years)  #alternative #2 to calculate multiple (cagr)
```

CAGR can be used to compare investments with different investment horizons.

# Compound Returns & Geometric Mean Return

```python
multiple=(1+btc.Return).prod()  #alternative #3 to calculate multiple (compound daily returns)
multiple
```

```python
n=btc.Return.count()
n
```

```python
geo_mean=multiple**(1/n)-1   #Geometric mean return (daily)
geo_mean
```

```python
(1+geo_mean)**n #alternative #4 to calculate multiple (geometric mean)
```

Compound returns, CAGR & geometric mean return are closely related concepts.

```python
mu= btc.Return.mean()
mu
```

```python
(1+mu)**n #calculate multiple? not possible with arithmetic mean!
```

The arithmetic mean return is always greater than the geometric mean return ... and less useful.
